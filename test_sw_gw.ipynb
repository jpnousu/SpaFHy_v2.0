#### Testing Di Ciacci et al., 2019 GW-SW exchange parameterization

```python
ditch_w = 1. # width of the channel
ditch_l = 20. # how is this done in Di Ciacci?
ditch_d = 40. # distance between ditches
dxys = np.array([20., 40., 80., 160.])
H_gw = 100. # gw head
H_sw = 90. # sw head
dt = 1. # timestep
M = 0.1 # thickness of channel bed [m]
Ksb = 1E-03 * 86400. # hydraulic conductivity of channel bed [m/d]
Khor = 1E-05 * 86400. # aquifer horizontal hydraulic conductivity [m/d]
Kver = 1E-06 * 86400. # aquifer vertical hydraulic conductivity [m/d]
Krad = np.sqrt(Khor*Kver) # aquifer radial hydraulic conductivity [m/d]
D = 3. # thickness of the saturated layer (i.e. aquifer) [m]
u = ditch_w * 1.2 # wetted perimeter [m]

for dxy in dxys:
    #ditch_l = dxy
    # resistances
    res_sb = M / Ksb * ditch_w * ditch_l # streambed resistance
    res_aqh = (3 * ditch_d * dxy - dxy**2) / (24 * ditch_d * Khor * D * ditch_l) # aquifer horizontal resistance
    res_aqr = (np.log(D/u) / (np.pi * Krad) * ditch_l) # aquifer radial resistance
    # flux
    S_dd = (H_gw - (H_sw) / (res_sb + res_aqh + res_aqr)) * dt / dxy**2
    print('S_dd =', S_dd)
    print('res_sb =', res_sb)
    print('res_aqh =', res_aqh)
    print('res_aqr =', res_aqr)
    print('')
```
